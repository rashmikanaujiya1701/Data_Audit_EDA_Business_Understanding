## Libraries Imports

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

plt.style.use("ggplot")
sns.set_palette("Set2")
```

## Load Data

```python
customers = pd.read_csv('/content/drive/MyDrive/T3_Project_Rashmi_Kanaujiya/D2C_Customer_Churn_Intelligence/Notebooks/Data_Audit_EDA_Business_Understanding/data/customers.csv')
orders = pd.read_csv('/content/drive/MyDrive/T3_Project_Rashmi_Kanaujiya/D2C_Customer_Churn_Intelligence/Notebooks/Data_Audit_EDA_Business_Understanding/data/orders.csv')
tickets = pd.read_csv('/content/drive/MyDrive/T3_Project_Rashmi_Kanaujiya/D2C_Customer_Churn_Intelligence/Notebooks/Data_Audit_EDA_Business_Understanding/data/support_tickets.csv')
events = pd.read_csv('/content/drive/MyDrive/T3_Project_Rashmi_Kanaujiya/D2C_Customer_Churn_Intelligence/Notebooks/Data_Audit_EDA_Business_Understanding/data/web_events_snapshot.csv')
churn = pd.read_csv('/content/drive/MyDrive/T3_Project_Rashmi_Kanaujiya/D2C_Customer_Churn_Intelligence/Notebooks/Data_Audit_EDA_Business_Understanding/data/churn_labels.csv')
campaigns = pd.read_csv('/content/drive/MyDrive/T3_Project_Rashmi_Kanaujiya/D2C_Customer_Churn_Intelligence/Notebooks/Data_Audit_EDA_Business_Understanding/data/intervention_history.csv')
```

## Dataset Audit

```python
datasets = {
    "customers": customers,
    "orders": orders,
    "tickets": tickets,
    "events": events,
    "churn": churn,
    "campaigns": campaigns
}

for name, df in datasets.items():
    print("="*50)
    print(name.upper())
    print(df.shape)
    display(df.head())
    display(df.info())
```

## Data Quality Section
Missing Values

```python
missing_report = {}

for name, df in datasets.items():
    missing = df.isnull().sum()
    missing_report[name] = missing

    print(f"\n{name}")
    print((missing/len(df))*100)
```

Visualization

```python
msno.matrix(customers)
plt.show()
```

Duplicate Records

```python
for name, df in datasets.items():
    print(
        name,
        "duplicates:",
        df.duplicated().sum()
    )
```

## Outlier Detection

```python
numeric_cols = orders.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=orders[col])
    plt.title(col)
    plt.show()
```

## Date Consistency

```python
date_cols = ['order_date']

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])

orders.sort_values('order_date').head()
```

```python
orders[
    orders['order_date'] > pd.Timestamp.today()
]
```

## Data Joining
Customer Level Dataset

```python
customer_df = customers.merge(
    churn,
    on="customer_id",
    how="left"
)
```

## Order Aggregation

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])

order_summary = orders.groupby('customer_id').agg(
    total_orders=('order_id', 'count'),
    total_spent=('gross_amount', 'sum'),
    avg_order_value=('gross_amount', 'mean'),
    last_order=('order_date', 'max')
).reset_index()

order_summary.head()
```

## Ticket Aggregation

```python
ticket_summary = tickets.groupby(
    'customer_id'
).agg(
    total_tickets=('ticket_id','count')
).reset_index()
```

## Campaign Aggregation

```python
campaign_summary = campaigns.groupby('customer_id').size().reset_index(
    name='campaigns_received'
)

campaign_summary.head()
```

## Final Analytical Dataset

```python
df = customer_df.merge(
    order_summary,
    on='customer_id',
    how='left'
)

df = df.merge(
    ticket_summary,
    on='customer_id',
    how='left'
)

df = df.merge(
    campaign_summary,
    on='customer_id',
    how='left'
)
```
